```python
import requests
from bs4 import BeautifulSoup
import datetime

def fetch_wheat_mandi_prices():

    mandi_data = {
        'Commodity' : [],
        'Arrival Date' : [],
        'Variety' : [],
        'State' : [],
        'District' : [],
        'Market' : [],
        'Min Price' : [],
        'Max Price' : [],
        'Avg Price' : []
    }

    url = "https://www.commodityonline.com/mandiprices/wheat/madhya-pradesh/rewa"
    resp = requests.get(url)
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, "html.parser")

    # Find average price element
    table = soup.find_all(class_ = "main-table2")[0].find('tbody')

    rows = table.find_all("tr")
    length = len(rows)

    for row in rows:
        cols = row.find_all("td")
        if len(cols) >= 9:
            mandi_data['Commodity'].append(cols[0].get_text(strip=True))
            mandi_data['Arrival Date'].append(cols[1].get_text(strip=True))
            mandi_data['Variety'].append(cols[2].get_text(strip=True))
            mandi_data['State'].append(cols[3].get_text(strip=True))
            mandi_data['District'].append(cols[4].get_text(strip=True))
            mandi_data['Market'].append(cols[5].get_text(strip=True))
            mandi_data['Min Price'].append(cols[6].get_text(strip=True))
            mandi_data['Max Price'].append(cols[7].get_text(strip=True))
            mandi_data['Avg Price'].append(cols[8].get_text(strip=True))

    return mandi_data
        



mandi_data = fetch_wheat_mandi_prices()
```

```python
import requests

def fetch_weather_rewa_weatherapi(api_key):
    city = "Rewa"
    url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}&aqi=no"
    
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()
    
    # Parse the relevant info
    weather_data = {
        "Location": data['location']['name'],
        "Region": data['location']['region'],
        "Country": data['location']['country'],
        "Local Time": data['location']['localtime'],
        "Temperature (C)": data['current']['temp_c'],
        "Condition": data['current']['condition']['text'],
        "Humidity (%)": data['current']['humidity'],
        "Wind (kph)": data['current']['wind_kph'],
        "Feels Like (C)": data['current']['feelslike_c'],
    }
    
    return weather_data

# --- Example Usage ---
API_KEY = "2434f2fa3ddf4180ad1223244250607"  # Replace with your actual WeatherAPI key
weather_info = fetch_weather_rewa_weatherapi(API_KEY)

# Pretty print
for k, v in weather_info.items():
    print(f"{k}: {v}")
```

```python
import requests

def fetch_5day_weather_forecast(api_key, city="Rewa", days=3):
    url = f"http://api.weatherapi.com/v1/forecast.json?key={api_key}&q={city}&days={days}&aqi=no&alerts=no"
    response = requests.get(url)
    response.raise_for_status()
    data = response.json()

    forecast_data = []
    for day in data['forecast']['forecastday']:
        forecast_data.append({
            "Date": day['date'],
            "Condition": day['day']['condition']['text'],
            "Avg Temp (°C)": day['day']['avgtemp_c'],
            "Max Temp (°C)": day['day']['maxtemp_c'],
            "Min Temp (°C)": day['day']['mintemp_c'],
            "Rainfall (mm)": day['day']['totalprecip_mm'],
            "Humidity (%)": day['day']['avghumidity']
        })

    return forecast_data

# Example Usage
API_KEY = "2434f2fa3ddf4180ad1223244250607"
forecast_info = fetch_5day_weather_forecast(API_KEY, city="Rewa", days=3)

# Pretty print
for day in forecast_info:
    print("\n".join([f"{k}: {v}" for k, v in day.items()]))
    print("-" * 40)
```

```python
import json

# Save to JSON
with open("weather_rewa.json", "w") as f:
    json.dump(forecast_info, f, indent=2)
```
